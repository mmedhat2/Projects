# 1 : Load Dataset




```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use(style='ggplot')

df = pd.read_csv('tmdb-movies.csv')
df.drop(columns=['id','homepage','tagline'],inplace=True)
df.sample(5)
```

# 2: Data Wrangling Process




```python
df.info()
```

## Changing the Format Of Dates

```python
df['release_date'] = pd.to_datetime(df['release_date'])
```

## Null Values in The Dataset

```python
#creating a data with the null values count
nullcount = df.isna().sum()
nullcount = pd.DataFrame(nullcount).reset_index()
#droping columns with no null values
nullcount = nullcount[nullcount[0] != 0]


#visulization for the nnull counts 
fig = plt.figure(figsize=(10, 5))
ax = plt.bar(nullcount['index'],nullcount[0])
ticks = plt.xticks(rotation = 45)
title = fig.suptitle("Number Of Null Values")
plt.ylabel('No. of Nulls')
plt.xlabel('Column')

print(nullcount)
```

## Number of zeros in each column

```python
zeros = pd.DataFrame()
nzero = []
cols = []
for i in df.columns:
    count = (df[i] == 0).sum()
    nzero.append(count)
    cols.append(i)

zeros['columns'] = cols
zeros['count'] = nzero
zeros
```

## Dropping Rows With Multiple Missing Values

```python

indexMissing = df[ (df['revenue'] == 0) & (df['budget'] == 0) & (df['genres'] == 0)].index
df.drop(indexMissing,inplace=True)
```

## Replacing  the Missing Values

```python

# I tried getting the gross and budget from the imdb website using webscraping but the code takes a long time to run



import urllib.request as urllib2
from bs4 import BeautifulSoup
import regex as re


hdr = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.11 (KHTML, like Gecko) Chrome/23.0.1271.64 Safari/537.11',
       'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
       'Accept-Charset': 'ISO-8859-1,utf-8;q=0.7,*;q=0.3',
       'Accept-Encoding': 'none',
       'Accept-Language': 'en-US,en;q=0.8',
       'Connection': 'keep-alive'}


#getting rows with no revenue
nobudgetdf = df[df['budget'] == 0] 



tags = []


#making a for loop to retreive the revenue for each link
try:
    for i in nobudgetdf.iloc[:3,1]:
        url= f"http://www.imdb.com/title/{i}/"


        req = urllib2.Request(url, headers=hdr)
        page = urllib2.urlopen(req)
        content = page.read()
        soup = BeautifulSoup(content, "html.parser")
        budgets_element = soup.find_all('label',class_="ipc-metadata-list-item__list-content-item")
        #appending all the html tags
        tags.append(budgets_element)

except:
    pass
        
  
```

```python
#filtering the retreived tags

budgets = []
for tag in tags:
    pucket = []
    for line in tag:
        budget = line.text.strip()
        if budget.startswith("$"):
            pucket.append(budget)
    budgets.append(pucket[0])
```

```python
#getting the main genre of the movie
df['genres'] = df['genres'].str.split('|').str[0]
```

## Successfull Movies

```python
TopMoviesdf = df.where(df['revenue'] > df['budget']*1.5)
TopMoviesdf.dropna(inplace=True)
```

```python
#Directors With Most Succefull Movies


#Top 10
sdirectors = TopMoviesdf['director'].value_counts().sort_values(ascending=False).reset_index()
topDirectors = sdirectors.iloc[0:10]

plt.figure(figsize=(8, 5))

plt.bar(topDirectors['index'],topDirectors['director'])
plt.xticks(size=8,rotation = 60,ha='right')
plt.xlabel('Movie')
plt.suptitle("Most Succefull Directors")
plt.ylabel('No. Of Movies')
```

## Percentage of Successfull movies To each director

```python
directorpercent = pd.merge(directorcount,sdirectors,on='index')
directorpercent.rename(columns={'index':'director','director_x':'movies','director_y':'successfull movies'},inplace=True)
directorpercent['percentage'] = directorpercent['successfull movies']/directorpercent['movies'] *100
directorpercent['percentage'] = directorpercent['percentage'].astype(int)
directorpercent.iloc[:20,:]
```

## the Importance of Each Column to the Revenue

```python
numeric_df = df.select_dtypes(include=['number'])
numeric_df = numeric_df[(numeric_df['budget'] != 0) & (numeric_df['revenue'] != 0)]


def plot_relation(i, x):
    sns.regplot(x=i,y='revenue',data=x,color='blue',scatter_kws={'alpha':0.3},line_kws={'color': 'red'})
    plt.xlabel(i)
    plt.ylabel('Revenue')

for i in numeric_df.columns:
    plot_relation(i, numeric_df)
    plt.show()
```

# Model Prediction

## predict if the movie will be succefull using the budget and other columns
 We Will Use the Data With With Budget and Revenue Then Predict The Data With No Revenue

```python
#import libraries
from sklearn import preprocessing
from sklearn.model_selection import train_test_split 
from xgboost import XGBRegressor
from sklearn import metrics 
from scipy import stats
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
```

## handling Categorical Columns

```python
#changing object columns to distinct values
obj = df.select_dtypes(include='object')
for i in obj.columns:
    le1 = LabelEncoder()
    df[i] = le1.fit_transform(df[i])
```

## Correlation Between Columns

```python
x = df.corr()
fig = plt.figure(figsize=(12, 5))
sns.heatmap(x)
```

```python
#features
modeldf = df[(df['budget']!= 0)&(df['revenue']!=0)]

x=modeldf[['popularity', 'budget','director','runtime', 'genres','release_year','budget_adj']]
y=modeldf['revenue']
x
```

```python
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size=0.3)

regr = XGBRegressor()
regr.fit(x_train, y_train)
regr.score(x_train,y_train)
```

```python
#regression visulization
test_predicted = regr.predict(x_test)

plt.figure(figsize=(7, 5))

sns.regplot(x=test_predicted,y=y_test,color='red')
```

## Predicting The Values With No Revenue

```python
predictdf = df[(df['budget'] != 0)&(df['revenue'] == 0) ]
```

```python
#selecting the columns to predict with

features = predictdf[['popularity', 'budget','director','runtime', 'genres','release_year','budget_adj']]
features
```

```python
prediction = regr.predict(features)
```

```python
predictdfcopy = predictdf.copy()
predictdfcopy.loc[:,'revenue'] = prediction
```
